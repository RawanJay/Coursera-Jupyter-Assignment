```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Define the URL
url = 'https://finance.yahoo.com/quote/TSLA/'

# Send the HTTP request
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Print the first 500 characters of the parsed HTML to check if it's working
print(soup.prettify()[:500])
```

```python
# Find the table in the HTML
table = soup.find('table')  # Adjust the selector if needed
if table:
    # Print out the first few rows of the table to see if it's correct
    print(table.prettify()[:500])
else:
    print("Table not found.")
```

```python
# Try finding all tables on the page
tables = soup.find_all('table')

# Print the number of tables found and some content from each
for i, table in enumerate(tables):
    print(f"Table {i} preview:")
    print(table.prettify()[:500])
    print("\n")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for Tesla
tsla = yf.Ticker("TSLA")
tsla_data = tsla.history(period="max")

def make_graph(stock_data, title1, title2):
    plt.figure(figsize=(14, 10))
    
    
    plt.subplot(2, 1, 1)  
    plt.plot(stock_data.index, stock_data['Close'], label='Close Price', color='blue')
    plt.title(title1)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    

    plt.subplot(2, 1, 2)  
    plt.plot(stock_data.index, stock_data['Volume'], label='Volume', color='orange')
    plt.title(title2)
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.legend()
    
    plt.tight_layout()  
    plt.show()


make_graph(tsla_data, 'Tesla Stock Price Over Time', 'Tesla Stock Volume Over Time')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for GameStop
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Define the make_graph function to handle subplots vertically stacked
def make_graph(stock_data, title1, title2):
    plt.figure(figsize=(14, 10))
    
    # First subplot: Closing Price
    plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st plot
    plt.plot(stock_data.index, stock_data['Close'], label='Close Price', color='blue')
    plt.title(title1)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    
    # Second subplot: Volume
    plt.subplot(2, 1, 2)  # 2 rows, 1 column, 2nd plot
    plt.plot(stock_data.index, stock_data['Volume'], label='Volume', color='orange')
    plt.title(title2)
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.legend()
    
    plt.tight_layout()  # Adjusts spacing between plots
    plt.show()

# Plot GameStop's stock price and volume
make_graph(gme_data, 'GameStop Stock Price Over Time', 'GameStop Stock Volume Over Time')
```
